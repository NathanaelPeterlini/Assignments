```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from urllib.request import urlopen
from bs4 import BeautifulSoup

%matplotlib inline
```

```python
url = "https://sport.orf.at/tabellen/stories/2288062/"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
text = soup.get_text()
print(text)
```

```python
soup.find_all('a')
```

```python
## first I wanted to locate all the different given adresses
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all('tr')
print(rows)
```

```python
# now I want to visualize the classification
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(36)
```

```python
drivers = df[0].str.split(',', expand=True)
drivers.head(20)
```

```python
drivers[0] = drivers[0].str.strip('[')
drivers.head(20)
```

```python
drivers.rename (columns = {0:'place', 1:'driver', 2:'Nationality', 3:'Team', 4: 'points'}, inplace = True)
print (drivers.head(20))
```
